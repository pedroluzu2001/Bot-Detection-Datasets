```python
df = pd.read_csv("C:\\Users\\Pedroluzuriaga\\OneDrive - KARNATAKA SA\\Escritorio\\Bots_actualizado\\tweeets_no_sentiment.csv")

print("Shape:", df.shape)
df.head()
```

```python
def clean_text(text):
    if pd.isna(text):
        return ""
    text = text.strip()
    if len(text) < 5:
        return ""
    return text

df["clean_content"] = df["content"].apply(clean_text)
```

```python
INTENT_LABELS = [
    "attack",
    "defense",
    "propaganda",
    "mockery",
    "call to action",
    "neutral"
]
```

```python
classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli",
    device=DEVICE
)
```

```python
sample_text = df.loc[df["clean_content"] != "", "clean_content"].iloc[0]

classifier(
    sample_text,
    INTENT_LABELS,
    multi_label=False
)
```

```python
def infer_intent(text):
    if not isinstance(text, str) or text.strip() == "":
        return {label: 0.0 for label in INTENT_LABELS}

    result = classifier(
        text,
        INTENT_LABELS,
        multi_label=False
    )

    return dict(zip(result["labels"], result["scores"]))
```

```python
import os

CHUNK_SIZE = 1000

results = []

for i in range(0, len(df), CHUNK_SIZE):
    chunk_file = f"intent_chunk_{i}.csv"

    # Si el chunk ya existe, lo cargamos y lo saltamos
    if os.path.exists(chunk_file):
        print(f"Saltando chunk {i} (ya existe)")
        results.append(pd.read_csv(chunk_file))
        continue

    print(f"Procesando chunk {i}")

    chunk = df.iloc[i:i+CHUNK_SIZE].copy()

    intent_scores = chunk["clean_content"].progress_apply(infer_intent)
    intent_df = pd.DataFrame(intent_scores.tolist())

    chunk = pd.concat(
        [chunk.reset_index(drop=True), intent_df],
        axis=1
    )

    chunk.to_csv(chunk_file, index=False)
    results.append(chunk)

df = pd.concat(results, ignore_index=True)

df.head()
```
